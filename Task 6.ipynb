Step 1 : Importing necessary libraries


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn import tree 
```

Step 2: Loading the Iris dataset

Dataset = "Iris.csv" (https://bit.ly/3kXTdox)

```python
data = pd.read_csv("Iris.csv")
data.head()
```

Step 3 : Data Exploration and Preprocessing

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

Step 4 : Extracting the features and target values

```python
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
X = data.loc[:, features].values
```

```python
y = data.Species
```

Step 5 : Splitting the data into training and testing datasets

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

Step 6 : Intialising and training the model with relevant model parameters

max_depth = 2, random_state = 42

```python
model = DecisionTreeClassifier(max_depth=2, random_state=0)
```

```python
model.fit(X_train, y_train)
```

Step 8 : Predicting using the model and evaluating its performance

```python
y_pred = model.predict(X_test)
```

```python
model.score(X_test, y_test)
```

```python
print(classification_report(y_test, y_pred))
```

Step 9 : Plotting the confusion matrix for better visualisation

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure()
sns.heatmap(cm, cmap = 'Blues', annot=True,square=True, linewidths= .5)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.tick_params(labelsize = 15)
```

Step 10 : Checking the optimum value of max_depth for our model

```python
max_depth_range = range(1,6)
accuracy = []

for depth in max_depth_range:
    model = DecisionTreeClassifier(max_depth=depth, random_state=0)
    model.fit(X_train, y_train)
    sc = model.score(X_test, y_test)
    accuracy.append(sc)
```

```python
plt.figure()

plt.plot(max_depth_range, accuracy)
```

Visualising the Decision Tree Classifier for max_depth = 5
